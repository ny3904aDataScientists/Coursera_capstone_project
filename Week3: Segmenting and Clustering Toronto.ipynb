### Let us explore all neighborhoods in Toronto Sub dataset

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    venues_list=[]
    nearbyvenues=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        url='https://api.foursquare.com/v2/venues/search?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, VERSION, lat, lng, radius, LIMIT)
        results=requests.get(url).json()
        surrounding_venues=results['response']['venues']
        v=json_normalize(surrounding_venues)
        v1=v.apply(get_category_type,axis=1)
        v=pd.DataFrame(v1,columns=['categories'])

        ### venues_list.append((v['categories']))
        ### print (venues_list)
        
        for c in v['categories']:
            venues_list.append((name,lat,lng,c))    
        
        
        
        for a_tup in venues_list:
            nearbyvenues.append(list(a_tup))


        data_frame_venues = pd.DataFrame(nearbyvenues)
        data_frame_venues.columns = ['Neighborhood', 'Neighborhood_Latitude', 'Neighborhood_Longitude', 'Venue_Category']                  
       
    
    return (data_frame_venues)
```

```python
Toronto_venues=getNearbyVenues(names=Y5['Neighbourhood'], latitudes=Y5['Latitude'],longitudes=Y5['Longitude'])
Toronto_venues.head()
```

```python
print('There are {} uniques categories.'.format(len(Toronto_venues['Venue_Category'].unique())))
```

```python
Toronto_venues_onehot = pd.get_dummies(Toronto_venues[['Venue_Category']], prefix="", prefix_sep="")
Toronto_venues_onehot['Neighborhood'] = Toronto_venues['Neighborhood'] 

fixed_columns = [Toronto_venues_onehot.columns[-1]] + list(Toronto_venues_onehot.columns[:-1])

Toronto_venues_onehot = Toronto_venues_onehot[fixed_columns]
Toronto_venues_onehot.head()
```

```python
Toronto_grouped = Toronto_venues_onehot.groupby('Neighborhood').mean().reset_index()
Toronto_grouped.head()
```

```python
from sklearn.cluster import KMeans
kclusters = 5
Toronto_grouped_clustering = Toronto_grouped.drop('Neighborhood', 1)
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Toronto_grouped_clustering)
kmeans.labels_[0:10] 
Toronto_grouped_clustering.head()
```

```python

Toronto_grouped_clustering.insert(0, 'Cluster Labels', kmeans.labels_)

Toronto_grouped_clustering['Neighbourhood']=Toronto_grouped['Neighborhood']
Toronto_grouped_clustering.head()
```

### Five Clusters

```python
Toronto_clustering=pd.merge(Toronto_grouped_clustering, Y5, how='left', on='Neighbourhood')


Toronto_clustering.sort_values(by=['Cluster Labels'])
```
